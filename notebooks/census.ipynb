```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
census = pd.read_csv('../data/census_davidson.csv')
census.head()
```

```python
census.info()
```

```python
census
```

```python
census_df = census.transpose()
census_df
```

```python
census_df = census_df.reset_index()
```

```python
census_df
```

```python
zip_codes = []

for x in census_df['index']:
    v = re.findall(r'Label', x)
    y = re.findall(r'United States', x)
    z = re.findall(r'\d{5}', x)
    if v:
        zip_codes.append(v)
    if y:
        zip_codes.append(y)
    if z:
        zip_codes.append(z)

zip_codes
```

```python
zip_codes = pd.DataFrame(zip_codes, columns = ['zip_codes'])
zip_codes
```

```python
census_df = pd.merge(census_df, zip_codes, left_index = True, right_index = True)
census_df
```

```python
census_df.info()
```

```python
census_clean.columns = census_clean.iloc[0]
census_clean
```

```python
census_clean = census_clean.rename(columns = {'Label': 'Zip Codes'})
census_clean
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
census = pd.read_csv('../data/census_davidson_2021.csv')
census
```

```python
census_df = census.transpose()
census_df = census_df.reset_index()
census_df.columns = census_df.iloc[0]
census_df = census_df.drop(labels = 0, axis = 0)
census_df = census_df.reset_index()
census_df = census_df.drop(labels = [0, 1, 2], axis = 0)
census_df = census_df.reset_index()
census_df = census_df.drop(labels = ['level_0', 'index'], axis = 1)
```

```python
zip_codes = []

for x in census_df['Label (Grouping)']:
    z = re.findall(r'\d{5}', x)
    if z:
        zip_codes.append(z)

zip_codes
```

```python
zip_codes = pd.DataFrame(zip_codes, columns = ['zip_codes'])
zip_codes
```

```python
census_df = pd.merge(census_df, zip_codes, left_index = True, right_index = True)
census_df
```
